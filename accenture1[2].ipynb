```python
def create_cost(n,cota):
    if n in prime:
        return c[n][1];
    else:
        #resta + división
        min = cota;
        for p in prime:
            a = n % p;
            ctemp = c[a][1] + c[p][1];
            if min - ctemp >=0:
                n1 = (n-a)//p;
                if n1 != 1:
                    ctemp += create_cost(n1,min-ctemp);
                if ctemp < min :
                    min= ctemp;
                    
                    #print (p)
            ctemp = c[p-a][1] + c[p][1];
            
            if min - ctemp >=0:
                n1 = (n-a)//p +1;
                if n1 != 1:
                    ctemp += create_cost(n1,min-ctemp);
                if ctemp < min :
                    min= ctemp;
                    #print (p)
        return min;
```
